```python
import math 

x = []
y = []

def findPeriod(length):
    pi = 3.14159265359
    gravity = 9.81
    return(2 * pi * (math.sqrt(length / gravity)))

i = 20
while i < 2000: 
    x.append(i)
    y.append(findPeriod(i))
    i += 20

```

```python
import math
import numpy as np 
from matplotlib import pyplot as plt 


plt.title("Galileo's Pendulum") 
plt.xlabel("length") 
plt.ylabel("period") 
plt.plot(x,y) 
plt.show()
```
